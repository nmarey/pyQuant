```python
import numpy as np
import pandas as pd
import yfinance as yf #pip install yfinance
from itertools import chain
# tick = yf.Ticker('^GSPC')
tick = yf.Ticker('SPY')
hist = tick.history(period="5Y", rounding=True)
```

```python
hist.head()
```

```python
colnames = list(range(int(round(min(hist['Low']),0)), 
                      (int(round(max(hist['High']),0))+1)))
```

```python
dcolname = list([["Dates"], colnames])
dcolname = list(chain.from_iterable(dcolname))
dcolname
```

```python
df = pd.DataFrame(columns=dcolname)
df
```

```python
df.Dates = hist.index.date
df = df.fillna(0)

for i in range(0,len(hist)):
    df.loc[df['Dates']== hist.index[i].date(), int(round(hist.Low[i],0)):int(round(hist.High[i],0))] = 1
    
df
```

```python
test = df[colnames].sum(axis=0).copy()
```

```python
test2 = pd.DataFrame(data=test, index=test.index, columns=["Close"])
```

```python
test2
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
k_models = []

size = min(11, len(test2))
for i in range(1, size):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(test2)
    wcss.append(kmeans.inertia_)
    k_models.append(kmeans)
    
saturation_point=0.05

optimum_k = len(wcss)-1
for i in range(0, len(wcss)-1):
    diff = abs(wcss[i+1] - wcss[i])
    if diff < saturation_point:
        optimum_k = i
        break

print("Optimum K is " + str(optimum_k + 1))

supp_lines = [item for sublist in k_models[optimum_k].cluster_centers_ for item in sublist]
```
